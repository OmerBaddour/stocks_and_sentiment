```python
import datetime
```

```python
today = datetime.date.today()
start = (today - datetime.timedelta(days=30)).strftime("%Y-%m-%d")
end = today.strftime("%Y-%m-%d")

tcs = yf.download('TCS',start,end)
```

```python
tcs
```

```python
type(tcs)
```

```python
tcs['Open'].plot(label = 'TCS', figsize = (15,7))
plt.title('Stock Prices of TCS')
```
